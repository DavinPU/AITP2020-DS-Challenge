```python
from ggplot import *
```

# Create a dataframe for values we are interested in, remove NaN values, and remove heights greater than 400

```python
import pandas as pd
df = pd.read_csv('/home/glentner/public/datasets/fiadb/TREE.csv', usecols=['HT'])
df = df.dropna()
df1 = df[(df.HT > 0) & (df.HT < 400)]
df2 = df[(df.HT > 400)]
df3 = df[(df.HT > 300) & (df.HT < 400)]
```

# Create a histogram for height

```python
a = ggplot(df1, aes(x='HT'))
a + geom_histogram(bins=20,
                   alpha=0.75,
                   color='black',
                   fill='lightblue',
                   xlab='Height',
                   ylab='Frequency',
                   title='Histogram of Tree Height')
```

# Create a dataframe with height and diameter

```python
diameterdf = pd.read_csv('/home/glentner/public/datasets/fiadb/TREE.csv', usecols=['DIA', 'HT', 'SPCD'])
diameterdf = diameterdf.dropna()
diameterdf1 = diameterdf[(diameterdf.HT > 0) & (diameterdf.HT < 400) & (diameterdf.DIA > 0) & (diameterdf.DIA < 75)]
diameterdf2 = diameterdf[(diameterdf.HT > 0) & (diameterdf.HT < 400) & (diameterdf.DIA > 40)]
```

# Create a histogram for diameter

```python
a = ggplot(diameterdf1, aes(x='DIA'))
a + geom_histogram(bins=20,
                   alpha=0.75,
                   color='black',
                   fill='lightgreen',
                   xlab='Height',
                   ylab='Frequency',
                   title='Histogram of Tree Diameter')
```

# Create a scatterplot of diameter vs height

```python
c = ggplot(diameterdf1, aes(x='DIA', y='HT'))
c + geom_point(alpha=0.6, color='green')
```
